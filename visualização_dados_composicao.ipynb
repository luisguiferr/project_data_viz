# Análise 1

Como está disposto o PIB nos Estados brasileiros no ano de 2020? É possível notar os Estados com maior e menor participação no PIB nacional?

### Tratando os dados

```python
# Criação de df com os dados desejados
df_pib_2020 = df_pib.copy()
df_pib_2020 = df_pib_2020.query('ano == 2020').set_index('sigla_uf').sort_values('pib',ascending=False)[['pib']]

df_pib_2020.head()
```

### Gerando gráfico de pizza

```python
# Geração de gráfico com os parâmetros para a visualização dos dados
fig, ax = plt.subplots(figsize=(6,6))
ax.pie(data=df_pib_2020, x='pib', labels=df_pib_2020.index, autopct='%.1f%%',
       radius=1.1, pctdistance=0.8, textprops={'size':12})

# Personalização do gráfico
plt.suptitle('Produto Interno Bruto (PIB) por estado (2020)', size=18, color=CINZA1)
plt.title('Em porcentagem (%)', fontsize=14, color=CINZA3, pad=15, x=0.09, y=0.99)

# Anotações do gráfico
ax.text(1.1, 0.3,
        'Um $\\bf{gráfico\\;de\\;pizza}$, consegue trazer a informação de\n'
        'participação no PIB de forma direta, no entanto, percebemos\n'
        'que o uso de $\\bf{muitas\\;fatias}$ atrapalham na análise dos\n'
        'dados para estados com $\\bf{menor}$ parcela no PIB.\n'
        'Além de gerar incômodo visualmente para quem lê.\n\n'
        'Percebemos que os estados de $\\bf{São\\;Paulo}$, $\\bf{Rio\\;de\\;Janeiro}$ e\n'
        '$\\bf{Minas\\;Gerais}$ possuem $\\bf{maior}$ participação no PIB nacional,\n'
        'porém, não conseguimos precisar quais estados tem a menor\n'
        'e quanto seria em percentual.',
        fontsize=12,
        linespacing=1.45,
        color=CINZA3, transform= ax.transAxes)

plt.show()
```

### Ajustando a visualização para vários dados (treemap)

```python
# Criando df com os dados desejados
pib_2020 = df_pib.copy()
pib_2020 = pib_2020.query('ano == 2020')[['regiao', 'sigla_uf', 'pib']]

# Ajustando os valores do PIB para bilhões e ordenando
pib_2020['pib'] = round(pib_2020['pib'] / 1e9, 0).astype('int64')
pib_2020 = pib_2020.sort_values('pib', ascending=False)

# Gerando coluna para a porcentagem da representação de cada estado no PIB
pib_2020['pib_%'] = pib_2020['pib'].div(pib_2020['pib'].sum(), axis=0)
pib_2020['pib_%'] = round(pib_2020['pib_%'] * 100, 2)

pib_2020.head()
```

```python
# Gerando o gráfico de árvore (TREEMAP)
fig = px.treemap(pib_2020, path=[px.Constant('Distribuição do PIB'), 'sigla_uf'],
                 values='pib_%',
                 color='regiao',
                 custom_data=['regiao', 'pib', 'pib_%'],
                 title='Distribuição do PIB nos estados brasileiros no ano de 2020 (em bilhões de reais)',
                 color_discrete_map={'(?)': BRANCO,
                                     'Sudeste':AZUL3,
                                     'Sul': LARANJA1,
                                     'Nordeste': AZUL5,
                                     'Centro-Oeste': VERDE1,
                                     'Norte': CINZA5})

# Ajustando o layout do gráfico
fig.update_layout(width=1400, height=600, margin=dict(t=50, l=0, r=0, b=25),
                  font_family='Arial',
                  font_size=14,
                  font_color=CINZA2,
                  title_font_color=CINZA1,
                  title_font_size=24)

fig.update_traces(marker=dict(cornerradius=5), texttemplate='<b>%{label}</b>',
                  hovertemplate='<b>Estado:</b> %{label}<br>'
                  '<b>Região</b> = %{customdata[0]}<br>'
                  '<b>PIB</b> = %{customdata[1]} (em bilhões)<br>'
                  '<b>Participação</b> = %{customdata[2]}%<br>')

fig.show()
```

# Análise 2


Houve uma significativa mudança na distribuição do PIB por região comparando os valores de 2002 e 2020?

### Tratando os dados

```python
# Criação do df com os dados desejados
df_pib_2002 = df_pib.query('ano == 2002')[['regiao', 'pib']]
df_pib_2020 = df_pib.query('ano == 2020')[['regiao', 'pib']]

df_pib_2002 = df_pib_2002.groupby('regiao')[['pib']].sum().sort_values('pib', ascending=False)
df_pib_2020 = df_pib_2020.groupby('regiao')[['pib']].sum().sort_values('pib', ascending=False)

display(df_pib_2002, df_pib_2020)
```

### Gerando gráfico de pizza

```python
# Geração de gráfico com os parâmetros para a visualização dos dados
fig, axs = plt.subplots(1, 2, figsize=(14,6))

# Definindo as cores do gráfico
cores = [AZUL3, LARANJA1, AZUL5, VERDE1, CINZA5]

# Personalização do gráfico
plt.suptitle('Produto Interno Bruto (PIB) por região',
             fontsize=18,
             color=CINZA1,
             ha='right',
             x=0.5,
             y=1.05)

# Gráfico 1
ptc, text, _ = axs[0].pie(data=df_pib_2002, x='pib', labels=df_pib_2002.index,
                          autopct='%.1f%%',
                          pctdistance=0.6,
                          textprops={'size':12,
                                     'fontweight':'bold',
                                     'color': BRANCO},
                          colors=cores)

# Título do gráfico 1
axs[0].annotate('Em 2002', xy=(50,350), size=12, xycoords='axes points',
                bbox=dict(boxstyle='round', fc=BRANCO, ec=CINZA3))

# Ajuste das cores das regiões relativas as fatias
for i, p in enumerate(ptc):
    text[i].set_color(p.get_facecolor())

# Gráfico 2
ptc, text, _ = axs[1].pie(data=df_pib_2020, x='pib', labels=df_pib_2020.index,
                          autopct='%.1f%%',
                          pctdistance=0.6,
                          textprops={'size':12,
                                     'fontweight':'bold',
                                     'color': BRANCO},
                          colors=cores)

# Título do gráfico 2
axs[1].annotate('Em 2020', xy=(50,350), size=12, xycoords='axes points',
                bbox=dict(boxstyle='round', fc=BRANCO, ec=CINZA3))

# Ajuste das cores das regiões relativas as fatias
for i, p in enumerate(ptc):
    text[i].set_color(p.get_facecolor())

axs[0].annotate('A Região Sudeste foi a única que\nreduziu a participação no PIB',
                xy=(180, 245), xycoords='axes points', xytext=(280, 295), textcoords='axes points',
                bbox=dict(boxstyle="round", fc=BRANCO, ec=CINZA3),
                size=10, arrowprops=dict(arrowstyle="->", fc=CINZA1, connectionstyle="arc,angleA=-90,angleB=0,armA=23,armB=30,rad=5"))

axs[1].annotate('\n\n', xy=(140, 245), xycoords='axes points', xytext=(-50, 295),
                textcoords='axes points', size=10, arrowprops=dict(arrowstyle="->", fc=CINZA1, connectionstyle="arc,angleA=-90,angleB=0,armA=30,armB=-30,rad=5"))

plt.show()
```

### Questionamento 2 e tratamento dos dados

Qual a participação da Região Norte nos valores agregados da indústria de todo Brasil nos anos de 2010 e 2020? Podemos notar algum comportamento?

Como está distribuído o PIB nos 3 últimos quinquênios (lustro) dos dados (2010, 2015, 2020) na Região Sul do Brasil divididos pelos Estados (Paraná, Santa Catarina e Rio Grande do Sul)?

### Tratando os dados

```python
# Selecionando os dados de acordo com o desejado
anos = [x for x in range(2010,2021,5)]

pib_sul = df_pib.query('regiao == "Sul" & ano == @anos')[['ano', 'sigla_uf', 'pib']]

pib_sul['pib'] = round(pib_sul['pib']/1e9, 2)
pib_sul = pib_sul.reset_index(drop=True)

pib_sul
```

```python
# Criação de tabela cruzada (crosstab) com os valores do pib de cada ano por região
pib_sul_cross = pd.crosstab(index=pib_sul.ano,
                            columns=pib_sul.sigla_uf,
                            values=pib_sul.pib,
                            aggfunc='sum',
                            normalize='index')

pib_sul_cross = pib_sul_cross.reset_index()

pib_sul_cross
```

```python
# Função para escrita de texto utilizando diferentes cores
# Adapado de https://github.com/empathy87/storytelling-with-data
def texto_colorido(x, y, texto, cores, esp=20, ax=None, **kw):
    cores = list(reversed(cores))
    t = ax.transData
    canvas = ax.figure.canvas

    for i, linha in enumerate(reversed(texto.split('\n'))):
        frases = linha.split('||')
        for s, cor in zip(frases, cores[i]):
            texto = ax.text(x, y, s, color=cor, transform=t, **kw)
            texto.draw(canvas.get_renderer())
            ex = texto.get_window_extent()
            t = transforms.offset_copy(texto._transform, x=ex.width,
                                       units='dots')

        t = transforms.offset_copy(ax.transData, x=0, y=(i + 1) * esp, units='dots')
```
